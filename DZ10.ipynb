```python
# Импортируем необходимые библиотеки
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Загружаем данные
from google.colab import files
uploaded = files.upload()

# Читаем загруженный CSV-файл
df = pd.read_csv("BostonHousing.csv")
```

```python
# Выводим первые 5 строк для ознакомления
print(df.head())
```

```python
# Анализ зависимости преступности от близости к реке
plt.figure(figsize=(6, 4))
sns.barplot(x=df["chas"], y=df["crim"], estimator=lambda x: x.mean(), hue=df["chas"], palette="Blues")
plt.xlabel("Граница с рекой Чарльз (0 - нет, 1 - да)")
plt.ylabel("Средний уровень преступности")
plt.title("Уровень преступности в зависимости от реки Чарльз")
plt.show()
```

```python
# Анализ зависимости числа комнат от доступности автострад
plt.figure(figsize=(6, 4))
sns.barplot(x=df["rad"], y=df["rm"], estimator=lambda x: x.mean(), hue=df["rad"], palette="Greens")
plt.xlabel("Индекс доступности автострад (rad)")
plt.ylabel("Среднее число комнат")
plt.title("Связь доступности автострад и числа комнат")
plt.show()
```

```python
# Анализ зависимости налога от уровня загрязнения воздуха
plt.figure(figsize=(6, 4))
sns.barplot(x=pd.cut(df["nox"], bins=5), y=df["tax"], estimator=lambda x: x.mean(), hue=pd.cut(df["nox"], bins=5), palette="Reds", legend=False)
plt.xticks(rotation=45)
plt.xlabel("Уровень загрязнения воздуха (группы)")
plt.ylabel("Средний налог")
plt.title("Связь загрязнения воздуха и налогообложения")
plt.show()
```
